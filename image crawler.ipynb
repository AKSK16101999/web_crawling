#  image crawler by using BeautifulSoup and requests.

```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
r=requests.get("https://in.images.search.yahoo.com/yhs/search;_ylt=AwrwJTSX0tZfEzQABX0O9olQ?p=electricity+grid+damage+images&fr=yhs-Lkry-newtab&hsimp=yhs-newtab&hspart=Lkry&imgty=photo&fr2=p%3As%2Cv%3Ai")
```

```python
r
```

```python
soup2=BeautifulSoup(r.text,"html.parser")
```

```python
soup2
```

```python
links=[]
```

```python
a=soup2.select('img[src^="https://tse2.mm.bing.net"]')
```

```python
for img in a:
    links.append(img['src'])
```

```python
for l in links:
    print(l)
```

```python
os.mkdir('elec_data')
i=1
for index,img_link in enumerate(links):
    if i<=200:
        img_data=requests.get(img_link).content
        with open("elec_data/"+str(index+1)+'.jpg','wb+')as f:
            f.write(img_data)
        i+=1
    else:
        f.close()
        break
```
